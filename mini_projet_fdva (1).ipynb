```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv("/content/fifa_eda_stats.csv")

# Select top players
players = df[df["Name"].isin(["L. Messi", "Cristiano Ronaldo", "Neymar Jr"])]

# Use correct attribute names from your dataset
attributes = ["Dribbling", "Finishing", "ShortPassing", "Acceleration", "ShotPower"]

# Get values
values = players[attributes].values
labels = players["Name"].values
num_attrs = len(attributes)

# Prepare angles for radar chart
angles = np.linspace(0, 2 * np.pi, num_attrs, endpoint=False).tolist()
values = np.concatenate((values, values[:, [0]]), axis=1)
angles += angles[:1]

# Plot radar chart
plt.figure(figsize=(7,7))
for i in range(len(labels)):
    plt.polar(angles, values[i], label=labels[i])
plt.xticks(angles[:-1], attributes)
plt.title("Top Player Skill Comparison (Radar Chart)")
plt.legend(loc="upper right", bbox_to_anchor=(1.2, 1.05))
plt.show()
```

```python
#  FIFA Player Performance Visualization
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/fifa_eda_stats.csv")

# ==============================
# 1 Scatter Plot: Age vs Overall Rating
# ==============================
plt.figure(figsize=(8,5))
plt.scatter(df["Age"], df["Overall"], alpha=0.6, color="orange", edgecolors="k")
plt.title("Age vs Overall Rating")
plt.xlabel("Age")
plt.ylabel("Overall Rating")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

# ==============================
# 2 Bar Chart: Top 10 Countries by Player Count
# ==============================
top_countries = df["Nationality"].value_counts().head(10)
plt.figure(figsize=(9,5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.title("Top 10 Countries by Number of Players")
plt.xlabel("Number of Players")
plt.ylabel("Country")
plt.show()

# ==============================
# 3 Radar Chart: Top Player Skill Comparison
# ==============================
# Select top players
players = df[df["Name"].isin(["L. Messi", "Cristiano Ronaldo", "Neymar Jr"])]

# Define skill attributes (adjust if needed based on dataset columns)
attributes = ["Dribbling", "Finishing", "ShortPassing", "Acceleration", "ShotPower"]

# Extract values and labels
values = players[attributes].values
labels = players["Name"].values
num_attrs = len(attributes)

# Create angles for radar chart
angles = np.linspace(0, 2 * np.pi, num_attrs, endpoint=False).tolist()
values = np.concatenate((values, values[:, [0]]), axis=1)
angles += angles[:1]

# Plot radar chart
plt.figure(figsize=(7,7))
for i in range(len(labels)):
    plt.polar(angles, values[i], label=labels[i])
plt.xticks(angles[:-1], attributes)
plt.title("Top Player Skill Comparison (Radar Chart)")
plt.legend(loc="upper right", bbox_to_anchor=(1.2, 1.05))
plt.show()
```
